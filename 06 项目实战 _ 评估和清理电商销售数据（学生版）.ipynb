# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘**畅销产品**，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 第一步、读取数据
导入数据分析所需要的库，并通过pandas的read_csv函数将原始数据文件"e_commerce.csv"里的数据内容，解析为DataFrame并赋值给变量original_data。 

```python
import pandas as pd
```

```python
original_data = pd.read_csv("e_commerce.csv")
```

```python
original_data.head()
```

## 第二步、评估数据
在这一部分，我们将对上一部分建立的`original_data`这个`DataFrame`所包含的数据进行评估。  
评估主要从两个方面进行，结构和内容，即整齐度和干净度。  
结构性问题指的是不符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”这三个标准。  
数据的内容性问题包含丢失数据、重复数据、无效数据等。

### 二-1 评估数据整齐度

```python
original_data.sample(10)
```

从抽样10行数据来看，符合“每列是一个变量，每行是一个观察值，每个单元格是一个值”，具体来看，每行是一次交易，每列是交易的各个变量，不存在结构性问题。

### 二-2 评估数据干净度

```python
original_data.info()
```

从输出结果来看，共有541909条观察值，`Description`、`CustomerID`存在缺失值。  
此外，`InvoiceDate`的数据类型应为日期，`CustomerID`的数据类型应为字符串，应进行格式转换。

#### 二-2-1 评估缺失数据

##### 二-2-1-1筛选`Description`缺失的行

```python
original_data[original_data["Description"].isnull()]
```

观察到`UnitPrice`为0，进一步确认是否均为0  
找`Description`为0的同时，是否有`UnitPrice`不为0的数据

```python
original_data[(original_data["Description"].isnull())&(original_data["UnitPrice"]!=0)]
```

输出结果为空，即`Description`为0时，`UnitPrice`也为0。   
`Description`表示产品名称，`UnitPrice`表示产品单价。数据分析目标为挖掘畅销产品，二者同时缺失时，可以认为数据无法提供有效信息，因此后续可删除。

##### 二-2-1-2筛选`CustomerID`缺失的行

```python
original_data[original_data["CustomerID"].isnull()]
```

`CustomerID`表示客户编号，不是分析畅销产品的必要变量，从结果来看，有些`CustomerID`缺失的交易数据仍然有效，因此保留此变量为空的观察值。

##### 二-2-1-3评估重复数据

有三个数据具有“唯一”性：   
InvoiceNo: 发票号码。6位数，作为交易的**唯一**标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。  
StockCode: 产品代码。5位数，作为产品的**唯一**标识符。  
CustomerID: 客户编号。5位数，作为客户的**唯一**标识符。
但，实际分析可知：   
每张发票可能对应多个产品，不同交易可能包含相同产品，同一顾客可能多次交易，因此均可以重复。   
针对此数据集，无需评估重复数据。

##### 二-2-1-4评估不一致数据（同一对象有不同名称）

可能存在不同值指代同一国家的情况

```python
original_data["Country"].value_counts()
```

UK、U.K.、United Kingdom 其实都是英国，但被分开统计了。  
USA 和 United States 都是美国。  
导致计数分散，影响后续分析的准确性。  
因此应该在清洗步骤对数值进行统一，只保留一个指代值。

##### 二-2-1-5评估无效或错误数据

使用`DataFrame`的`describe`方法，对数值统计信息进行快速了解

```python
original_data.describe()
```

从输出结果看，`Quantity`和`UnitPrice`存在负数，对后续数值分析产生影响。  
因此，依次筛选出数值为负数的观察值，进一步评估其意义。

###### 二-2-1-5-1评估`Quantity`负数值

```python
original_data[original_data["Quantity"]<0]
```

观察到Quantity负数的发票订单目前可见是C开头，即交易被取消，进一步验证是否所有均是如此。  
看看是否存在Quantity为负数，但InvoiceNO不以C开头的观察值。

```python
original_data[(original_data["Quantity"]<0)&(original_data["InvoiceNo"].str[0]!="C")]
```

发现有一些Quantity为负的订单，不是退货订单。   
但观察到，猜测UnitPrice为0，进一步验证是否属实。  
Q为负+订单不是C+U不为0，是否还有观察值

```python
original_data[(original_data["Quantity"]<0)&(original_data["InvoiceNo"].str[0]!="C")&(original_data["UnitPrice"]!=0)]
```

无数据，这说明，Quantity为负的订单，要么是C开头（被取消了）；要么是价格为0。  
均不设销售有效数据。  
在清理数据时可以删除。

###### 二-2-1-5-2评估`UnitPrice`负数值

```python
original_data[original_data["UnitPrice"]<0]
```

结果可见，为“坏账调整”，不属于实际商品交易，在数据清洗中删除。

## 第三步、清理数据

根据前面评估部分得到的结论，我们需要进行的数据清理包括：
- 把 `InvoiceDate` 变量的数据类型转换为日期时间
- 把 `CustomerID` 变量的数据类型转换为字符串
- 把 `Description` 变量缺失的观察值删除
- 把 `Country` 变量值 `"USA"` 替换为 `"United States"`
- 把 `Country` 变量值 `"UK"`、`"U.K."` 替换为 `"United Kingdom"`
- 把 `Quantity` 变量值为负数的观察值删除
- 把 `UnitPrice` 变量值为负数的观察值删除

为了区分清洗后的数据与原始数据，创建新变量`cleaned_data`，让它成为`original_data`复制出的副本，之后的清理步骤都在`cleaned_data`上。

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

### 三-1 把`InvoiceData`变量的数据类型转化为时间日期类型：

```python
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data['InvoiceDate']
```

### 三-2 把CustomerID 变量的数据类型转换为字符串：

```python
cleaned_data['CustomerID'] = cleaned_data["CustomerID"].astype(str)
cleaned_data['CustomerID']
```

删除CustomerID结尾的.0（保留第一个到倒数第二个字符）

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].str.slice(0,-2)
cleaned_data['CustomerID']
```

### 三-3 把 Description 变量缺失的观察值删除:

```python
cleaned_data.dropna(subset=["Description"], inplace=True)
```

验证是否删除：

```python
cleaned_data["Description"].isnull().sum()
```

### 三-4 把`Country`变量值`"USA"`替换为`"United States"`，并检查替换后`"USA"`变量值个数：

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"USA": "United States"})
```

```python
len(cleaned_data[cleaned_data["Country"] == "USA"])
```

### 三-5 把`Country`变量值`"UK"`、`"U.K."`替换为`"United Kingdom"`，并检查替换后`"UK"`和`"U.K."`变量值个数：

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"UK": "United Kingdom", "U.K.": "United Kingdom"})
print(len(cleaned_data[cleaned_data["Country"] == "UK"]))
print(len(cleaned_data[cleaned_data["Country"] == "U.K."]))
```

### 三-6 把`Quantity`变量值为负数的观察值删除，并检查替换后`Quantity`变量值为负数的个数：

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >= 0]
```

```python
len(cleaned_data[cleaned_data["Quantity"] < 0])
```

### 三-7 把`UnitPrice`变量值为负数的观察值删除，并检查替换后`UnitPrice`变量值为负数的个数：

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]
```

```python
len(cleaned_data[cleaned_data["UnitPrice"] < 0])
```

完成数据清理后，把干净整齐的数据保存到新的文件里，文件名为`e_commerce_cleaned.csv`。

看一眼老文件，发现索引无实际含义，所以后续index=False（否则默认索引保存为单独的一列）
